# Feature Engineering

```python
# Define categories for the mood variable
categories = ['sad', 'mellow', 'upbeat', 'happy']

data['mood'] = pd.cut(
    x=data['valence'],
    bins=[0, 0.25, 0.5, 0.75, 1.0],
    labels=categories
)

# Add new categories to the mood variable
data['mood'] = data['mood'].cat.add_categories(['high energy', 'fast-paced'])
data.loc[data['energy'] >= 0.7, 'mood'] = 'high energy'
data.loc[data['tempo'] >= 120, 'mood'] = 'fast-paced'

# Print the value counts of each mood category
sns.countplot(data=data, x='mood', palette="mako")
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
# Select a subset of features for genre identification
features_temp = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
data_temp =data[features_temp]

data_temp['tempo_variance'] = data_temp['tempo'].rolling(window=10).var()
data_temp['speechiness_variance'] = data_temp['speechiness'].rolling(window=10).var()
```

```python
# Drop any rows with missing values
data_temp = data_temp.dropna()
```

```python
# Use KMeans clustering to identify genre clusters
X = data_temp.values
kmeans = KMeans(n_clusters=5, random_state=0).fit(X)
data_temp['genre'] = kmeans.labels_
```

```python
# Map genre cluster labels to genre names
genre_names = {0: 'Rock', 1: 'Pop', 2: 'Hip Hop', 3: 'Electronic', 4: 'Other'}
data_temp['genre_name'] = data_temp['genre'].apply(lambda x: genre_names[x])
```

```python
# Print out the resulting genre names for each song
sns.countplot(data=data_temp, y='genre_name', palette="mako")
plt.show()
```

```python
def remove_outliers(data,c):
  q1=data[c].quantile(0.25)
  q3=data[c].quantile(0.75)
  IQR=q3-q1
  lp=q1-1.5*IQR
  up=q3+1.5*IQR
  data[c]=np.where(data[c]<lp,data[c].median(),data[c])
  data[c]=np.where(data[c]>up,data[c].median(),data[c])
  return data
```

```python
from sklearn.neighbors import NearestNeighbors

```

```python

nn_model = NearestNeighbors(metric='cosine', algorithm='brute')
```

```python
def recom(input_value):
    # Determine the column to filter based on the input
    filter_columns = ["artists", "track_name", "album_name","track_genre"]

    tfidf_matrix = tfidf_vectorizer.fit_transform(data_reduce['features'])

    nn_model.fit(tfidf_matrix)

    for filter_column in filter_columns:
        filtered_data = data_reduce[data_reduce[filter_column] == input_value]
        if not filtered_data.empty:
            # Get the index of the first entry for the given input
            input_index = filtered_data.index[0]

            # Use Nearest Neighbors to find similar songs
            _, indices = nn_model.kneighbors(tfidf_matrix[input_index], n_neighbors=20)

            # Exclude the input song itself
            indices = indices.flatten()[1:]

            # Create a set to keep track of unique track names
            unique_track_names = set()

            # Print the recommended unique track names
            for index in indices:
                current_track_name = data_reduce.iloc[index]['track_name']
                if current_track_name not in unique_track_names:
                    print(current_track_name)
                    unique_track_names.add(current_track_name)
            return

```
